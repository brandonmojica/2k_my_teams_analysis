```python
def split_athleticism(x):


    athletic_array = []
    athletic_array_2 = []
    counter = 0
    length = (len(x) + 1)
    
    while counter < length:
            
           
        Athleticism = (re.split('Athleticism', x['athleticism'][counter]))
                #athletic_array.append(Athleticism)

        Speed = (re.split('Speed',Athleticism[1]))
                #athletic_array.append(Speed)

        Acceleration = re.split('Acceleration',Speed[1])
                #athletic_array.append(Acceleration)

        Speed = (re.split('Speed',Athleticism[1]))
                #athletic_array.append(Speed)

        Vertical = re.split('Vertical',Acceleration[1])
                #athletic_array.append(Vertical)

        Strength = re.split('Strength',Vertical[1])
                #athletic_array.append(Strength)

        Stamina = re.split('Stamina',Strength[1])
                #athletic_array.append(Stamina)

        Hustle = re.split('Hustle',Stamina[1])
                #athletic_array.append(Hustle)

        Overall_durability = re.split('Overall durability',Hustle[1])
                #athletic_array.append(Overall_durability)

        counter+=1
            
        y = list([Athleticism[0],Speed[0],Acceleration[0],Vertical[0],Strength[0],Stamina[0],Hustle[0],Overall_durability[0]])
        athletic_array_2.append(y)
    return np.asarray(athletic_array_2)
```

```python
Athleticism_array = split_athleticism(player_data)
```
